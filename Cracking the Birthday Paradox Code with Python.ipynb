```python
import random
import matplotlib.pyplot as plt

def birthday_paradox(n, seed=None):
    """Simulate the birthday paradox with the Mersenne Twister algorithm"""
    random.seed(seed)
    birthdays = [random.randint(1, 365) for i in range(n)]
    return len(set(birthdays)) != len(birthdays)

def estimate_probability(n, num_simulations=1000):
    """Estimate the probability of two people having the same birthday"""
    count = 0
    for i in range(num_simulations):
        if birthday_paradox(n, seed=i):
            count += 1
    return count / num_simulations

# Example usage
n_values = list(range(2, 101))
probs = [estimate_probability(n) for n in n_values]

plt.plot(n_values, probs)
plt.xlabel("Number of People")
plt.ylabel("Probability of Two People Having the Same Birthday")
plt.title("Birthday Paradox Simulation Results")
plt.show()
```

```python
import numpy as np
from scipy.stats import randint
import random

def birthday_paradox_random(n):
    """Simulate the birthday paradox using the random module"""
    birthdays = [np.random.randint(1, 365) for _ in range(n)]
    return len(set(birthdays)) != len(birthdays)

def estimate_probability_random(n, num_simulations=1000):
    """Estimate the probability of two people having the same birthday using the random module"""
    count = 0
    for i in range(num_simulations):
        if birthday_paradox_random(n):
            count += 1
    return count / num_simulations

def birthday_paradox_numpy(n):
    """Simulate the birthday paradox using the numpy.random module"""
    birthdays = np.random.randint(1, 365, size=n)
    return len(set(birthdays)) != len(birthdays)

def estimate_probability_numpy(n, num_simulations=1000):
    """Estimate the probability of two people having the same birthday using the numpy.random module"""
    count = 0
    for i in range(num_simulations):
        if birthday_paradox_numpy(n):
            count += 1
    return count / num_simulations

def birthday_paradox_scipy(n):
    """Simulate the birthday paradox using the scipy.stats.randint function"""
    birthdays = randint.rvs(1, 365, size=n, random_state=np.random)
    return len(set(birthdays)) != len(birthdays)

def estimate_probability_scipy(n, num_simulations=1000):
    """Estimate the probability of two people having the same birthday using scipy.stats.randint"""
    count = 0
    for i in range(num_simulations):
        if birthday_paradox_scipy(n):
            count += 1
    return count / num_simulations

# Example usage
n_values = list(range(2, 101))
probs = [estimate_probability_random(n) for n in n_values]
probs_numpy = [estimate_probability_numpy(n) for n in n_values]
probs_scipy = [estimate_probability_scipy(n) for n in n_values]

import matplotlib.pyplot as plt

plt.plot(n_values, probs, label="random")
plt.plot(n_values, probs_numpy, label="numpy.random")
plt.plot(n_values, probs_scipy, label="scipy.stats.randint")
plt.xlabel("Number of people")
plt.ylabel("Probability of two people having the same birthday")
plt.legend()
plt.show()
```
